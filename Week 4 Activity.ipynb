```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
data=pd.read_csv('Telco-Customer-Churn.csv')
data.head()
```

# Compare churn count with respect to gender

```python
plt.figure(figsize=(8,9))
sns.countplot(x ='Churn', data = data, hue='gender')
plt.title('Churn count compared to Gender', fontsize=15)
```

# Find out how many female senior citizens there in the dataset

```python

s=data[(data['SeniorCitizen'] == 1) & (data['gender'] == 'Female')]
plt.figure(figsize=(6,6))
sns.countplot(s['gender'], color='Green',)
plt.title('Total female senior citizens',fontsize=16)
plt.xlabel('Gender')
plt.ylabel('Count')
s['gender'].count()
```

Total no. of female senior citizen = 568

# Compare 'tenure' with 'Total Charges'

```python
plt.figure(figsize=(8,8))
plt.scatter(data['tenure'],data['TotalCharges'], s=6)
plt.title('Plot of Tenture VS Total Charges',fontsize=18)
plt.xlabel('Tenture')
plt.ylabel('Total Charges')
```

# Find out which contract is preferred by the senior citizen.

```python
c=data[(data['SeniorCitizen']==1)]
plt.figure(figsize =(10, 7))
c.Contract.value_counts(sort=True).plot.pie(fontsize=12,autopct ='%1.1f%%')
plt.legend()
plt.title('Contracts preferred by Senior Citizens',fontsize=18)
```

Month-to-Month contract is mostly preferred by senior citizens

```python
m=data[(data['SeniorCitizen']==1)]
plt.figure(figsize =(10,14))
m.PaymentMethod.value_counts(sort=True).plot.pie(fontsize=12,autopct ='%1.1f%%')
plt.legend()
plt.title('Payment method opted by Senior Citizens',fontsize=18)
```
